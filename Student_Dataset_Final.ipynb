```python
#Importing and Viweing the data
#Creation of number of dataframes as required for different analysis
import pandas as pd
df = pd.read_csv(r'/Users/shashankashekharsharma/Downloads/student_ml_dataset.csv')
df_female = df[df['Gender'] == 'Female']
df_male = df[df['Gender'] == 'Male']
df_new = df[['Name', 'Physics_Marks','Chemistry_Marks','Math_Marks']]
df_new['Total_Marks']=df_new['Physics_Marks']+df_new['Chemistry_Marks']+df_new['Math_Marks']
print(df)
print(df_new)
print(df_female)
print(df_male)
```

```python
# Analysis of Dataset involves:
#     1. Mean
#     2. Median
#     3. Percentile
#     4. Count of various variables
#     5. Finding totals
# There is also comparitive analysis on the basis of Gender
# A hypothesis is also being tested and various graphs have been plotted to which shows various insights of the student dataset
# Pandas, Matplotlib and Seaborn has been used
```

```python
#Means
print('Mean age is ',df['Age'].mean())
print('Mean hours studied is ',df['Hours_Studied'].mean())
print('Mean IQ is',df['IQ'].mean())
print('Mean Physics Marks is',df['Physics_Marks'].mean())
print('Mean Math Marks is ',df['Math_Marks'].mean())
print('Mean Chemistry marks is ',df['Chemistry_Marks'].mean())
print("Females")
print('Mean age of females is ',df_female['Age'].mean())
print('Mean hours studied by females is ',df_female['Hours_Studied'].mean())
print('Mean IQ of females is',df_female['IQ'].mean())
female_physics=df_female['Physics_Marks'].mean()
print('Mean Physics Marks of females is',female_physics)
female_math=df_female['Math_Marks'].mean()
print('Mean Math Marks of females is ',female_math)
female_chemistry=df_female['Chemistry_Marks'].mean()
print('Mean Chemistry marks of females is ',female_chemistry)
print("Males")
print('Mean age of males is ',df_male['Age'].mean())
print('Mean hours studied by males is ',df_male['Hours_Studied'].mean())
print('Mean IQ of males is',df_male['IQ'].mean())
male_physics=df_male['Physics_Marks'].mean()
print('Mean Physics Marks of males is',male_physics)
male_math=df_male['Math_Marks'].mean()
print('Mean Math Marks of males is ',male_math)
male_chemistry=df_male['Chemistry_Marks'].mean()
print('Mean Chemistry marks of males is ',male_chemistry)
```

```python
#Median
print('Median age is ',df['Age'].median())
print('Median hours studied is ',df['Hours_Studied'].median())
print('Median IQ is',df['IQ'].median())
print('Median Physics Marks is',df['Physics_Marks'].median())
print('Median Math Marks is ',df['Math_Marks'].median())
print('Median Chemistry marks is ',df['Chemistry_Marks'].median())
```

```python
#Comparitive Analysis of Gender
males = df['Gender'].value_counts()['Male']
females = df['Gender'].value_counts()['Female']
print("Number of Males: ",males)
print("Number of Females: ",females)
```

```python
#Highest and Lowest Marks
physics_highest=df_new['Physics_Marks'].max()
chemistry_highest=df_new['Chemistry_Marks'].max()
math_highest=df_new['Math_Marks'].max()
total_highest=df_new['Total_Marks'].max()
physics_lowest=df_new['Physics_Marks'].min()
chemistry_lowest=df_new['Chemistry_Marks'].min()
math_lowest=df_new['Math_Marks'].min()
total_lowest=df_new['Total_Marks'].min()
print("Highest Physics Marks ",physics_highest)
print("Highest Chemistry Marks ",chemistry_highest)
print("Highest Math Marks ",math_highest)
print("Highest Total Marks ",total_highest)
print("Lowest Physics Marks ",physics_lowest)
print("Lowest Chemistry Marks ",chemistry_lowest)
print("Lowest Math Marks ",math_lowest)
print("Lowest Total Marks ",total_lowest)
```

```python
#Ranges of various subjects and total
physics_range=physics_highest-physics_lowest
chemistry_range=chemistry_highest-chemistry_lowest
math_range=math_highest-math_lowest
total_range=total_highest-total_lowest
print("Range of Physics ",physics_range)
print("Range of Chemistry ",chemistry_range)
print("Range of Math ",math_range)
print("Range of Total marks ",total_range)
```

```python
#Standard Deviation and Variance of Various Subjects
std_phy=df_new['Physics_Marks'].std()
var_phy=df_new['Physics_Marks'].var()
print("Standard Deviation of Physics is ",std_phy)
print("Variance of Physics is ",var_phy)
std_chemistry=df_new['Chemistry_Marks'].std()
var_chemistry=df_new['Chemistry_Marks'].var()
print("Standard Deviation of Chemistry is ",std_chemistry)
print("Variance of Chemistry is ",var_chemistry)
std_math=df_new['Math_Marks'].std()
var_math=df_new['Math_Marks'].var()
print("Standard Deviation of Math is ",std_math)
print("Variance of Math is ",var_math)
```

```python
#Standard Deviation and Variance of Various Subjects only for Males
std_phy=df_male['Physics_Marks'].std()
var_phy=df_male['Physics_Marks'].var()
print("Standard Deviation of Physics (only males) is ",std_phy)
print("Variance of Physics (only males) is ",var_phy)
std_chemistry=df_male['Chemistry_Marks'].std()
var_chemistry=df_male['Chemistry_Marks'].var()
print("Standard Deviation of Chemistry (only males) is ",std_chemistry)
print("Variance of Chemistry (only males) is ",var_chemistry)
std_math=df_male['Math_Marks'].std()
var_math=df_male['Math_Marks'].var()
print("Standard Deviation of Math (only males) is ",std_math)
print("Variance of Math (only males) is ",var_math)
```

```python
#Standard Deviation and Variance of Various Subjects only for females
std_phy=df_female['Physics_Marks'].std()
var_phy=df_female['Physics_Marks'].var()
print("Standard Deviation of Physics (only females) is ",std_phy)
print("Variance of Physics (only females) is ",var_phy)
std_chemistry=df_female['Chemistry_Marks'].std()
var_chemistry=df_female['Chemistry_Marks'].var()
print("Standard Deviation of Chemistry (only females) is ",std_chemistry)
print("Variance of Chemistry (only females) is ",var_chemistry)
std_math=df_female['Math_Marks'].std()
var_math=df_female['Math_Marks'].var()
print("Standard Deviation of Math (only females) is ",std_math)
print("Variance of Math (only females) is ",var_math)
```

```python
#Students having part time jobs and not having part time jobs
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(
    x = "Has_Part_Time_Job",
    data=df
)
plt.xlabel("Has Part Time Job")
plt.ylabel("Number of Students")
plt.title("Number of Students with and without Part Time Jobs")
plt.show()
```

```python
#Hypothesis Testing
import scipy
from scipy import stats
print(stats.ttest_ind(df[df['Gender'] == 'Male']['IQ'], df[df['Gender'] == 'Female']['IQ']))
```

```python
#Coorelation between Hours Studied and IQ
print(df['IQ'].corr(df['Hours_Studied']))
```

```python
#Plot of relation between IQ scores and number of students
import matplotlib.pyplot as plt
import numpy as np

iq_scores = df['IQ'].values
plt.hist(iq_scores)
plt.xlabel("IQ Score")
plt.ylabel("Number of Students")
plt.title("Distribution of IQ Scores")
plt.show()
```

```python
#box plot of the IQ scores of males and females
import matplotlib.pyplot as plt
import seaborn as sns

sns.boxplot(
    x = "Gender",
    y = "IQ",
    showmeans=True,
    data=df
)
plt.xlabel("Gender")
plt.ylabel("IQ Score")
plt.title("Distribution of IQ Scores by Gender")
plt.show()
```

```python
#scatter plot of the IQ scores and hours studied
import matplotlib.pyplot as plt
import numpy as np
iq_scores = df['IQ'].values
hours_studied = df['Hours_Studied'].values
plt.scatter(hours_studied, iq_scores)
plt.xlabel("Hours Studied")
plt.ylabel("IQ Score")
plt.title("Relationship between IQ Scores and Hours Studied")
plt.show()
```

```python
#Heatmap
import seaborn as sns
import matplotlib.pyplot as plt

# Create a correlation matrix
sns.heatmap(df.corr(), annot=True)
plt.title("Correlation Matrix")
plt.show()
```
